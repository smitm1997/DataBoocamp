## Question 2

```python
nl = nl.set_index('name')
```

```python
nl
```

```python
nl['BatAverage'] = nl['Hits'] / nl['AtBats']
nl['HRAverage'] = nl['HR'] / nl['AtBats']

nl
```

```python
print('The mean batting average is', nl.BatAverage.mean())
```

```python
print('The median homerun average is', nl.HRAverage.median())
```

```python
print('The correlation between BatAverage and Wins is',nl['BatAverage'].corr(nl['Wins']))
```

```python
print('The correlation between Wins and Attendance is',nl['Wins'].corr(nl['attendance']))
```

```python
nl['BatAverage'].plot(kind='barh', xlim=(.22,.28), color='blue')
```

```python
%matplotlib inline
fig, ax = plt.subplots(nrows=1, ncols=2) #Part a

nl.plot(x='Wins', y='BatAverage', kind='scatter',ax=ax[0], color='green', s=nl['attendance']/50000)   # first plot, part b
ax[0].set_xlabel('Wins')
ax[0].set_ylabel('BatAverage')

nl.plot(x='Wins', y='HRAverage', kind='scatter',ax=ax[1], color='red', s=nl['attendance']/50000)     # second plot, part c
ax[1].set_xlabel('Wins')
ax[1].set_ylabel('HRAverage')

ax[0].set_title('Batting Average vs Wins') #part d
ax[1].set_title('Homerun Average vs Wins')
```
